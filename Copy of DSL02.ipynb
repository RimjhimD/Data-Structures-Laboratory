# Grade finder

```python
grade = int(input("Enter your number : "))

if grade >= 80 and grade <=100 :
  print("A+")
elif grade >= 75 and grade <80 :
  print("A")
elif grade >= 70 and grade <75:
  print("A-")
elif grade >= 65 and grade <70:
  print("B+")
elif grade >= 60 and grade <64:
  print("B")
elif grade >= 55 and grade <60:
  print("B-")
elif grade >= 50 and grade <55:
  print("C+")
elif grade >= 45 and grade <50:
  print("C")
elif grade >= 40 and grade <45:
  print("D")
elif grade >= 0 and grade <40:
  print("F")
else:
  print("Invalid input")
```

# Prime number Section

```python
prime_list = []
test = int(input("Enter a number: "))

for num in range(2, test):
    flag = True

    for i in range(2, int((test+1)/2) + 1):
        if num % i == 0:
            flag = False
            break

    if flag:
        prime_list.append(num)

print(prime_list)
```

# Ans

```python
numbers = []
n = int(input('Enter a range : '))
for i in range(2,n+1):
  status = 0
  for j in range(2,i):
    if i % j ==0:
      status = 1
      break
  if status==0:
    numbers.append(i)
print(numbers)
```
