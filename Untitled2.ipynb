Delete by value

```python
def linear_search(A,N,ITEM):
    pos = -1
    for i in range(N):
      if A[i] == ITEM:
         pos = i
         break
    return pos

def deletion(A,N,K):
  for i in range(K,N-1):
    A[i] = A[i+1]
  traversal(A,N-1)

A = [10,20,30,40,50]
N = len(A)
ITEM = 40
K = linear_search(A,N,ITEM)
if K == -1:
   print("Element not found!")
else:
   deletion(A,N,K)
```

Binary search

```python
def binary_search(A,N,S):
   st = 1
   end = N
   mid = int((st + end)/2)
   pos = -1

   while st <=end:
        if A[mid] == S:
           pos = mid
           break
        elif S < A[mid]:
             end = mid - 1
        elif S > A[mid]:
              st = mid + 1
         mid = int((st + end)/2)
    if pos == -1:
         print("Element not found!")
    else:
        return pos
 A = [10,20,30,40,50]
 N = len(A)
 S = 500
 loc = binary_search(A,N,S)



```
